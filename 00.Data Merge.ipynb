```python
import pandas as pd
```

```python
inpatient_raw = pd.read_csv("data/Train_Inpatientdata-1542865627584.csv")
```

```python
outpatient_raw = pd.read_csv("data/Train_Outpatientdata-1542865627584.csv")
```

```python
beneficiary_raw = pd.read_csv("data/Train_Beneficiarydata-1542865627584.csv")
```

```python
target_raw = pd.read_csv('data/Train-1542865627584.csv')
```

```python
# glance the data
display(target_raw.head())
display(beneficiary_raw.head())
display(inpatient_raw.head())
display(outpatient_raw.head())
```

```python
# Assign inpatient data as 1, outpatient data as 0
inpatient_raw['In_Out']=1
outpatient_raw['In_Out']=0
```

```python
inpatient_raw.columns
```

```python
outpatient_raw.columns
```

```python
# concat the inpatient and outpatient data
concat_df=pd.concat([inpatient_raw,outpatient_raw],axis=0)
concat_df.head()
```

```python
concat_df.columns
```

```python
concat_df.shape
```

```python
merge_bene_df=concat_df.merge(beneficiary_raw, on='BeneID', how='left')
merge_bene_df.head()
```

```python
merge_provider_df=merge_bene_df.merge(target_raw, on = 'Provider', how ='left')
merge_provider_df.drop_duplicates(inplace=True)
merge_provider_df
```

```python
#export the merged data to csv
merge_provider_df.to_csv('Sample_Data.csv')
```
